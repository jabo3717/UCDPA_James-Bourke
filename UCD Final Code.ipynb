```python
#merging
final_result=[]
for i in Result_df['Result']:
    if(i > 0):
        final_result.append('won')
    else:
        final_result.append('lost')
final_result
```

```python
from pandas import DataFrame
df = DataFrame (final_result,columns=['Final_Result'])
```

```python
Result_df.merge(df,right_index=True,left_index=True)
```

```python
#use functions to create reusable code and usage of numpy
import numpy as np
```

```python
# Mean is the arithmetic mean of all dataset
#average gives the average of the column
#median is the middle element of the sorted dataset
#Mode is the value of the dataset the occurs most frequently
def summary_data(col_name):
    me = np.mean(Result_df[col_name])
    av = np.average(Result_df[col_name])
    med = np.median(Result_df[col_name])
    mi = np.min(Result_df[col_name])
    ma = np.max(Result_df[col_name])
    print("The mean of the column", col_name,"is",me)
    print("The average of the column", col_name,"is",av)
    print("The median of the column", col_name,"is",med)
    print("The minimum value of the column", col_name,"is",mi)
    print("The maximum value of the column", col_name,"is",ma)
    
```

```python
summary_data('Result')
```

```python
summary_data('Opposition Rating')
```

```python
#Lists
home_venue = ['Belfast', 'Croke Park', 'Lansdowne Road','Dublin', 'Limerick']
total = 0
count = 0
for i in Result_df['Location']:
    if(i in home_venue):
        total = total+Result_df['Result'].iloc(i)
        count = count + 1
aver = total/count
print("The average results of the location belonging to the list home venue is ", aver)
        
```
